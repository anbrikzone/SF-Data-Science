# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
from itertools import combinations
import pandas as pd

def cast_combination(data):
    cast_list = []
    for pair in combinations(data,2):
        cast_list.append(list(pair))
    return cast_list

data = pd.read_csv('movie_bd_v5.csv')
data['cast'] = data.cast.apply(lambda x: x.split("|"))
data['cast'] = data.cast.apply(cast_combination)
data = data.explode('cast')
data['cast'] = data.cast.apply(lambda x: str(x))
data.cast.value_counts().sort_values(ascending=False)
```

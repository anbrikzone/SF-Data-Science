# 11. Какого жанра фильмов больше всего?

```python
answers['11'] = 'Drama'
```

```python
# эту задачу тоже можно решать разными подходами, попробуй реализовать разные варианты
# если будешь добавлять функцию - выноси ее в предобработку что в начале
q11_data = data

dict_genres = {}

# Create dictionary of genres
for str_genre in q11_data.genres:
    for genre in str_genre.split('|'):
        #print(genre)
        if genre in dict_genres:
            dict_genres[genre] = dict_genres[genre]+1
        else:
            dict_genres[genre] = 1

# Frequently genre in dataset            
often_genre = {k:v for k, v in dict_genres.items() if v == max(dict_genres.values())}
often_genre
```

ВАРИАНТ 2

```python
cnt = Counter()

# We also can use Counter to make Dictionary
for str_genre in q11_data.genres:
    for genre in str_genre.split('|'):
        #print(genre)
        if genre in dict_genres:
            cnt[genre] +=1
        else:
            cnt[genre] = 1

# Frequently genre in dataset using properties of Counter lib
cnt.most_common(1)
```

# 14. Какой режисер снял больше всего фильмов в стиле Action?

```python
answers['14'] = 'Robert Rodriguez'
```

```python
data = pd.read_csv('movie_bd_v5.csv')
q14_data = data
# To make divided by genres and directors
q14_data['genres'] = q14_data.genres.apply(lambda x: x.split("|"))
q14_data['director'] = q14_data.director.apply(lambda x: x.split("|"))
#Unroll list data using explode function
exdata = q14_data.explode('genres')
exdata = exdata.explode('director')
#looking for Action genre
exdata = exdata[exdata['genres']=='Action']
exdata.groupby('director').genres.count().sort_values(ascending=False)
```
